```python
import requests
from bs4 import BeautifulSoup as BS
from math import ceil
import concurrent.futures
import threading
from multiprocessing import Pool
import pandas as pd
from tqdm import tqdm
import time
import random
```

```python
class Etherscan():
    delay = 5
    userAgent = ['Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36',
                 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36',
                 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.113 Safari/537.36',
                 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
                 'Thunderstorm/1.0 (Linux)',
                 'Mozilla/5.0 (iPhone; CPU iPhone OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148',
                 'Mozilla/5.0 (iPhone; CPU iPhone OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/12.1 Mobile/15E148 Safari/604.1'
                ]
    
    def getTokensCount(self):
        time.sleep(self.delay)
        req = requests.get('https://etherscan.io/tokens-nft',
                       headers = {
                'User-Agent': random.choice(self.userAgent)
            })
        soup = BS(req.text, 'lxml')
        total = soup.find('i', {'id': 'spinwheel'}).parent.getText().split(' ')[3].replace(',', '')
        return int(total)

    def getTokensLinks(self, url):
        time.sleep(self.delay)
        req = requests.get(url,
                           headers = {
                    'User-Agent': random.choice(self.userAgent)
                })

        soup = BS(req.text, 'lxml')
        try:
            table = soup.find('table', {'id': 'tblResult'})
            rows = table.findAll('tr')
            tokens = []
            for tr in rows:
                cols = tr.findAll('td')
                for col in cols:
                    if col.find('a') != None:
                        #print(col.find('a').getText())
                        tokens.append(col.find('a').get('href'))
            return tokens
        except:
            pass

    def getTokensLinksMultiple(self, urls):
        TokensLinks = []
        with concurrent.futures.ThreadPoolExecutor(max_workers=12) as executor:
            linksList = list(tqdm(executor.map(self.getTokensLinks, urls), total=len(urls)))
            for links in linksList:
                if links != None:
                    for link in links:
                        TokensLinks.append('https://etherscan.io'+link)      
        return TokensLinks


    def getCode(self, url):
        time.sleep(self.delay)
        req = requests.get(url,
                            headers = {
                        'User-Agent': random.choice(self.userAgent)
                    })

        soup = BS(req.text, 'lxml')
        try:
            code = []
            code.append(soup.find('pre', {'id': 'editor1'}).getText())
            return code
        except:
            pass


    def getCodeMultiple(self, urls):
        ContractCode = []
        with concurrent.futures.ThreadPoolExecutor(max_workers=12) as executor:
            linksList = list(tqdm(executor.map(self.getCode, urls), total=len(urls)))
            for links in linksList:
                if links != None:
                    for link in links:
                        TokensLinks.append(link)      
        return ContractCode

    def generateDataset(self, url):
        contracts = []
        urls = []
        contracts.append(url.split('/')[-1])
        urls.append(url)
        url = 'https://etherscan.io/address/'+url.split('/')[-1]+'#code'
        code = self.getCode(url)
        d = {'Contract':contracts,'Link':urls, 'Code':code}
        df = pd.DataFrame(d)
        return df

    def generateDatasetMultiple(self, urls):
        with concurrent.futures.ThreadPoolExecutor(max_workers=12) as executor:
            df = pd.concat(tqdm(executor.map(self.generateDataset, urls), total=len(urls)))
        return df
```

```python
pages = ceil(getTokensCount()/100)
urls = ['https://etherscan.io/tokens-nft?ps=100&p='+str(x) for x in range(615, pages)]

eth = Etherscan()

print('[*] Getting TOP NFT tokens..')
TokensLinks = eth.getTokensLinksMultiple(urls)

print('[*] Generating dataset..')
df = eth.generateDatasetMultiple(TokensLinks)
```
