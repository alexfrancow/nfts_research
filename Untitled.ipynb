```python
import requests
from bs4 import BeautifulSoup as BS
from math import ceil
import concurrent.futures
import threading
from multiprocessing import Pool
import pandas as pd
from tqdm import tqdm
import time
import random
import numpy as np
import re
```

```python
class Etherscan():
    delay = 8
    userAgent = ['Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36',
                 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36',
                 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.113 Safari/537.36',
                 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
                 'Thunderstorm/1.0 (Linux)',
                 'Mozilla/5.0 (iPhone; CPU iPhone OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148',
                 'Mozilla/5.0 (iPhone; CPU iPhone OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/12.1 Mobile/15E148 Safari/604.1'
                ]
    ethereumNetwork = 'https://etherscan.io'
    maxWorkers = 12

    def getTokensCount(self):
        '''
        Get the total number of NFTs in the ethereum network
        :return: The number of tokens
        '''
        time.sleep(self.delay)
        req = requests.get(self.ethereumNetwork + '/tokens-nft',
                       headers = {
                'User-Agent': random.choice(self.userAgent)
            })
        soup = BS(req.text, 'lxml')
        total = soup.find('i', {'id': 'spinwheel'}).parent.getText().split(' ')[3].replace(',', '')
        return int(total)

    def getTokensLinks(self, url):
        '''
        Parse the etherscan main webpage to get the links of the NFTs projects
        :param url: The url of the project
        :return: The links of the NFTs
        '''
        time.sleep(self.delay)
        req = requests.get(url,
                           headers = {
                    'User-Agent': random.choice(self.userAgent)
                })

        soup = BS(req.text, 'lxml')
        try:
            table = soup.find('table', {'id': 'tblResult'})
            rows = table.findAll('tr')
            tokens = []
            for tr in rows:
                cols = tr.findAll('td')
                for col in cols:
                    if col.find('a') != None:
                        tokens.append({'Name':col.find('a').getText(),'Link':self.ethereumNetwork + col.find('a').get('href')})
                        #tokens.append(col.find('a').get('href'))
            return tokens
        except:
            pass

    def getTokensLinksMultiple(self, urls):
        '''
        Invoke the getTokensLinks function for multiple projects
        :param urls: The urls of the etherscan pages (list)
        :return: The links of the NFTs
        '''
        TokensLinks = []
        with concurrent.futures.ThreadPoolExecutor(max_workers=self.maxWorkers) as executor:
            linksList = list(tqdm(executor.map(self.getTokensLinks, urls), total=len(urls)))
            for links in linksList:
                if links != None:
                    for link in links:
                        TokensLinks.append(link)
        return TokensLinks


    def getCode(self, url):
        '''
        Parse the etherscan webpage to get the code, files and ABI of the project.
        :param url: The url of the NFT project
        :return: The code, files and ABI of the project
        '''
        time.sleep(self.delay)
        req = requests.get(url + '#code',
                            headers = {
                        'User-Agent': random.choice(self.userAgent)
                    })

        soup = BS(req.text, 'lxml')
        # find all the pre tag with id editor regex
        editor1 = [x.get('id') for x in soup.findAll('pre', {'id': re.compile('editor')})]

        code = []
        filesL = []
        filesS = ""
        abi = []

        # Get code if only have one file
        if len(editor1) == 1:
            y = soup.find('pre', {'id': editor1[0]})
            code.append(y.getText())
            f = y.find_previous_sibling('div').getText()
            filesL.append(f)

        # Get all files if have more than one file
        else:
            for x in editor1:
                y = soup.find('pre', {'id': x})
                f = y.find_previous_sibling('div').getText()
                if ':' in f:
                    filesS += str(f.split(': ')[1]) + ", "
                else:
                    filesS += str(f) + ", "

                # Get only the code of the project
                if x == 'editor1':
                    code.append(y.getText())
            filesL.append(filesS[:-2])
            
        # Fix errors
        if len(code) == 0:
            code.append('NaN')
        if len(filesL) == 0:
            filesL.append('NaN')

        # Get ABI
        try:
            abi.append(soup.find('pre', {'id': 'js-copytextarea2'}).getText())
        except:
            abi.append('NaN')

        try:
            contractDiv = soup.find("div", string="Contract Name:")
            contractName = contractDiv.find_next_sibling("div").getText().replace('\n', '')
        except:
            contractName = 'NaN'
        try:
            compilerDiv = soup.find("div", string="Compiler Version")
            compilerVersion = compilerDiv.find_next_sibling("div").getText().replace('\n', '')
        except:
            compilerVersion = 'NaN'
        try:
            webproject = soup.find("a", {"title" : "External Site - More Info"}).get('href')
        except:
            webproject = 'NaN'
        try:
            creator = soup.find("a", {"title" : "Creator Address"}).get('href')
        except:
            creator = 'NaN'

        time.sleep(self.delay)
        req = requests.get(url.replace('/address', '/token'),
                            headers = {
                        'User-Agent': random.choice(self.userAgent)
                    })

        soup = BS(req.text, 'lxml')
        try:
            officialSite = soup.find("div", string="Official Site:")
            officialSite = officialSite.find_next_sibling("div").getText()
        except:
            officialSite = 'NaN'

        return code, filesL, abi, contractName, compilerVersion, webproject, creator, officialSite

    def generateDataset(self, url):
        '''
        Generate the dataset that contains all NFT projects with their code, files and ABI.
        :param url: The url of the NFT project
        :return: The dataset of the project
        '''
        contracts = []
        urls = []
        names = []
        contracts.append(url['Link'].split('/')[-1])
        urls.append(url['Link'])
        names.append(url['Name'])
        url = self.ethereumNetwork + '/address/' + url['Link'].split('/')[-1]
        code, filesL, abi, contractName, compilerVersion, webproject, creator, officialSite = self.getCode(url)
        d = {'Name':names, 'Contract':contracts, 'Link':urls, 'ABI':abi, 'Code':code, 'Files':filesL, 'Compiler':compilerVersion, 'ContractName':contractName, 'Webproject':webproject, 'Creator':creator, 'officialSite':officialSite}
        df = pd.DataFrame(d)
        return df

    def generateDatasetMultiple(self, urls):
        '''
        Invoke the generateDataset function for multiple projects
        :param urls: The urls of the NFT projects (list)
        :return: The dataset of the projects
        '''
        with concurrent.futures.ThreadPoolExecutor(max_workers=self.maxWorkers) as executor:
            df = pd.concat(tqdm(executor.map(self.generateDataset, urls), total=len(urls)))
        return df
```

```python
# Function that returns function names from solana block code
import re
def solana(code):
    """
    :param code: solana block code
    :return: list of function names find with re
    """
    return re.findall(r'(?<=function )[a-zA-Z0-9_]+', code)

# Apply function to each row in the code column
df['SolanaFunctions'] = df['Code'].apply(lambda x: solana(x))
df

# Function that returns onlyOwner function names from solana block code
def solanaOnlyOwnerFunctions(code):
    """
    :param code: solana block code
    :return: list of public function names find with re
    """
    functions = re.findall(r'((?<=function ).+{)', code)
    onlyOwnerFunctions = []
    for function in functions:
        if "onlyOwner" in function:   
            onlyOwnerFunctions.append(re.findall(r'[^(]+', function)[0]) 
    return onlyOwnerFunctions

df['solanaOnlyOwnerFunctions'] = df['Code'].apply(lambda x: solanaOnlyOwnerFunctions(x))
df

# difference between solana functions and onlyOwner functions
df['solanaFunctionsDiff'] = df['SolanaFunctions'].apply(lambda x: list(set(x) - set(df[['solanaOnlyOwnerFunctions']])))
df
```

```python
# Search similar contracts code with IA
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Generate DF
df = pd.read_excel('data122_modified.xlsx')

# Remove duplicated rows
df.drop_duplicates('CodeHash256', inplace=True)

# remove solidity comments in block code
df['Code'] = df['Code'].apply(lambda x: x.split('//')[0])

# Vectorizer to convert a collection of raw documents to a matrix of TF-IDF features
vectorizer = TfidfVectorizer()

# Learn vocabulary and idf, return term-document matrix.
tfidf = vectorizer.fit_transform(df['Code'].values.astype('U'))

# Array mapping from feature integer indices to feature name
words = vectorizer.get_feature_names()

# Compute cosine similarity between samples in X and Y.
similarity_matrix = cosine_similarity(tfidf, tfidf)

# Matrix product
similarity_matrix

# Instead of using fit_transform, you need to first fit 
# the new document to the TFIDF matrix corpus like this:
queryTFIDF = TfidfVectorizer().fit(words)

# We can check that using a new document text
query = '''function mint( uint quantity, bytes calldata signature ) external payable {
    require( quantity <= MAX_ORDER,             "FishyFam: order too big"             );
    require( msg.value >= PRICE * quantity, "FishyFam: ether sent is not correct" );

    uint max = msg.sender == whalePass ? WHALE_WALLET : MAX_WALLET;
    require( _balances[ msg.sender ] + quantity <= max, "FishyFam: don't be greedy" );

    uint supply = totalSupply();
    require( supply + quantity <= MAX_SUPPLY, "FishyFam: mint/order exceeds supply" );

    if( isMainsaleActive ){
      //no-op
    }
    else if( isPresaleActive ){
      verifySignature( quantity.toString(), signature );
    }
    else{
      revert( "Sale is not active" );
    }

    unchecked{
      for(uint i; i < quantity; ++i){
        _mint( msg.sender, tokens.length );
      }
    }
  }
'''

# Now we can 'transform' this vector into that matrix shape by using the transform function:
queryTFIDF = queryTFIDF.transform([query])

# As we transformed our query in a tfidf object
# we can calculate the cosine similarity in comparison with 
# our pevious corpora
cosine_similarities = cosine_similarity(queryTFIDF, tfidf).flatten()

# Get most similar jobs based on next text
related_product_indices = cosine_similarities.argsort()[:-11:-1]
related_product_indices

# Get Name column of related jobs
df[['Name', 'CodeHash256']].iloc[related_product_indices]
```
